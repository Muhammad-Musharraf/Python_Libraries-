# Creating Series From List

```python
# string series
country=['Pakistan','Russia','India','Srilanka','Nepal','Maldives']

# Series have two thing----> index & value
series=pd.Series(country)
print(series)

# Note : pandas call the string data type as object
```

```python
# integer series

runs=[34,99,78,88,23,65]
s=pd.Series(runs)

print(s)
```

```python
#creating series with custom index
# value
marks=[65,88,97,67,55,49,90]

# index
subjects=['English','Science','Math','Chemistry','Computer Science','Urdu','Pst']

pd.Series(marks,index=subjects)
```

```python
# setting a name of series

mark=pd.Series(marks,index=subjects,name='Musharraf Markssheets')

print(mark)
```

```python
#dictionary
marks={'Urdu':67,'chemistry':63,'physics':98,'English':68,'islamiat':88,'Social studies':89,'Science':78,'computer science':54}

marks_series=pd.Series(marks, name='Musharraf Marksheets')

print(marks_series.name)
print(marks_series)
```

```python
marks_df = marks_series.to_frame()
# marks_df.columns = [marks_series.name]  # Set column name from Series name
marks_df
```

```python
# size ----> return total number of items in series (also count nan values)
marks_series.size
```

```python
# dtype ----> return the data type of series items
marks_series.dtype
```

```python
# name ----> return the name of series
marks_series.name
```

```python
# is_unique -----> Return True , if all items in series is unique . else false

print(marks_series.is_unique)
```

```python
print(pd.Series([1,2,3,4,3,2,5,6,7]).is_unique)
```

```python
# index -----> return all index
marks_series.index
```

```python
#values -----> return all values
marks_series.values
```

#  Series Methods


```python
# head -----> return starting 5 rows of series (by default)

print(marks_series.head())

print()

# can change with the number of rows given number
print(marks_series.head(6))
```

```python
# tail -----> return last 5 rows of series (by default)
print(marks_series.tail())

print()
# can change with the number of rows given number
print(marks_series.tail(3))
```

```python
# sample -----> return one row randomly from series (by default)
print(marks_series.sample())

print()
# can change with the number of samples given number
print(marks_series.sample(3))
```

```python
# value_counts -----> retrun total frequency of each value from descending to ascending
marks={'Urdu':67,'chemistry':63,'physics':78,'English':88,'islamiat':88,'Social studies':89,'Science':67,'computer science':54}
marks=pd.Series(marks)


marks.value_counts()
```

```python
# sort_values -----> sort all values in ascending (by default)
print(marks.sort_values())
```

```python
# change the order to descending
print(marks.sort_values(ascending=False))
```

```python
# sort_index -----> sort the index in ascending (by default)
print(marks.sort_index())
```

```python
# change the order to descending
print(marks.sort_index(ascending=False))
```

# Series Indexing & Slicing

```python
# integer indexing
print(marks)

#positive indexing
print(marks[5])

#Negative indexing
print(marks[-2])

#String Indexing
print(marks["physics"])

#String Indexing
print(marks["English"])
```

```python
# integer slicing

#positive sicing
print(marks[2:5])

# negative indexing
print(marks[-6:-2])


#string sicing
print(marks['physics':'computer science'])
```

```python
# Fancy indexing ----> multiple indexing values
marks[[1,3,5,7]]
```

# Editing Series

```python
print(marks)

#change values using integer index
marks[2]=90
print(marks)

#change values using integer index
marks['Science']=99
print(marks)
```

```python
# what if the called index item not exist -----> pandas add that as new item at the last

marks['Commerce']=65
print(marks)
```

```python
# it is also possible with slicing

marks[2:6]=[100,95,93,90]
print(marks)
```

```python
# it is also possible with fancy indexing

marks[[2,4,6]]=[1,0,1]
print(marks)
```

# Important Series Methods

```python
# astype ----> change data type
print(marks)
print(marks.astype('float'))
```

```python
#duplicated ---> return True if dublicates, else False

print(marks.duplicated())
print(marks.duplicated().sum())
```

```python
# drop_duplicates ----> remove the duplicates values
marks=marks.drop_duplicates()
print(marks)
```

```python
#making new series
x=pd.Series([1,2,np.nan,3,np.nan,np.nan,4,5,np.nan,6,np.nan,7,8,9,10])
print(x)
```

```python
# isnull ----> return true if null value else false
print(x.isnull())

# total sum
print(x.isnull().sum())
```

```python
# dropna -----> use to remove the null values
print(x.dropna())
```

```python
# fillna ----> use to fill nan values by your own logic

print(x.fillna(x.mean()))
```

```python
#isin ----> return True if value available, else false
print(x.isin([3]))
```

```python
 #more than one
print(x.isin([3,8,9]))
```

```python
#boolean indexing
print(x[x.isin([3,8,9])])
```
