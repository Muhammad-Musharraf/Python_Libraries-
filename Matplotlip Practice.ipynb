* # Subplot

A subplot is a feature that allows multiple plots to be displayed within a single figure. It
divides the figure into a grid of rows and columns, enabling you to visualize multiple
graphs side by side for easy comparison of different datasets or relationships.


```python
# multiple plots using grid, side by side ---> plotting (2x1) grid

#fig ---> store figure where graph plot
#ax ----> save array for plot

fig,ax=plt.subplots(nrows=2,ncols=1,figsize=(10,6))

ax[0].scatter(df['avg'],df['strike_rate'])

ax[1].scatter(df['avg'],df['runs'],color='red')

ax[0].set_title("Avg vs Strike Rate")
ax[0].set_ylabel("Strike Rate")



ax[1].set_title("Avg vs Runs")
ax[0].set_xlabel("Avg")
ax[0].set_ylabel("Runs")




fig
```

```python
#plotting (2x2) grid ----> 2D Array

fig , ax=plt.subplots(nrows=2,ncols=2,figsize=(10,10))
ax[0,0].scatter(df['avg'],df['strike_rate'])

ax[0,1].scatter(df['avg'],df['runs'],color='red')

ax[1,0].hist(df['avg'],color='yellow')

ax[1,1].hist(df['runs'],color='pink')

plt.show()





```

# * 3D Scatter Plots:
A 3D scatter plot is a graphical representation that shows the relationship between three
numerical variables in a three-dimensional space. Each point is plotted using its X, Y, and
Z coordinates, allowing you to visualize patterns, trends, and correlations among three
variables simultaneously.


```python
fig=plt.figure(figsize=(10,5))

ax=plt.subplot(projection='3d')

ax.scatter3D(df['avg'],df['strike_rate'],df['runs'])

ax.set_title('IPL Batsman analysis')

ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')
ax.set_zlabel('Runs')

plt.show()
```

# 3. Bar chart:

A bar chart is a graphical representation used to compare categorical and numerical
data using rectangular bars. The categorical variable is shown on the X-axis, and the
keyboard_arrow_down 3. Bar chart:
numerical variable on the Y-axis. The length or height of each bar represents the value
or frequency of that category, making it easy to compare different groups visually.
