## list vs Numpy
* FAST
* Less memory
* convenient

```python
# less Memory ---> numpy take less memory as compare to python list
import sys
py_list=list(range(10000))
np_array=np.arange(10000)

print("Python list size",sys.getsizeof(py_list),"Bytes")
print("Numpy array size",np_array.nbytes,"Bytes")
```

```python
# Faster ---> Numpy Operational speed much faster than python list
import time

py_list=list(range(1000000))
start=time.time() # current Time
py_result=[]
for x in py_list:
  py_result.append(x*2)
end=time.time()
print("Python list Time",end-start)

np_array=np.arange(1000000)
start=time.time()
np_result=np_array*2 # Vectorization
end=time.time()
print("Numpy array list",end-start)
```
