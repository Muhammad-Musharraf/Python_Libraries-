# Merge, concatenate and join - Pandas

```python
d1={'EEID':[101,102,103,104,105,106,107,108],
                  'Name':['Finnegan','Orion','Silvanus','Alaric','Griffith','Osias','Ulric','Basilio'],
                  'Age':[23,22,28,29,27,37,48,36]}


d2={'EEID':[109,100,111,104,105,106,107,108],
                  
'Salary':[4300,3200,4857,5900,6700,2709,3867,3500]}

df1=pd.DataFrame(d1)

df2=pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='EEID',how='inner')
```

```python
pd.merge(df1,df2,on='EEID',how='left')
```

```python
pd.merge(df1,df2,on='EEID',how='right')
```

# Pandas-Compare DataFrames

```python
dict={'Furits':['Mango','Banana','Orange','Apple'],
     'Price':[120,50,70,100],
     'Quality':[15,10,30,20]}


data=pd.DataFrame(dict)
```

```python
data
```

```python
data1=data.copy()
data1

data1.loc[0,'Price']=150

data1.loc[2,'Price']=100
data1.loc[3,'Price']=140

data1.loc[0,'Quality']=30
data1.loc[2,'Quality']=50
data1.loc[3,'Quality']=35

print(data1)
```

```python
data.compare(data1)
```

```python
print(data.compare(data1,align_axis=0))
```

```python
data.compare(data1,keep_equal=True)
```

```python
data.compare(data1,keep_shape=True)
```

# Pivoting and Melting Dataframes

```python
dic= {
    "keys": ["k1", "k2", "k1", "k2"],
    "Names": ["John", "Ben", "David", "Peter"],
    "Houses": ["red", "blue", "green", "red"],
    "Grade":["3rd","8TH","9TH","7th"]
}

df5=pd.DataFrame(dic)
df5
```

```python
df5.pivot(index="keys",columns="Names",values=["Houses","Grade"])
```

```python
df5
```

```python
print(pd.melt(df5,id_vars=["Names"],value_vars=["Houses"]))
```

```python
pd.melt(df5,id_vars=["Names"],value_vars=["Houses","Grade"])
```

```python
pd.melt(df5,id_vars=["Names"],value_vars=["Houses","Grade"],var_name="Houses&Grade",value_name="Values")
```
